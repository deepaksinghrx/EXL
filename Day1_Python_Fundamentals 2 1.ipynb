```python
import json
import os

INPUT_FILE = "customer.json"
OUTPUT_FILE = "report.txt"


def load_customers(INPUT_FILE):
    with open(INPUT_FILE, "r", encoding="utf-8") as source:
        return json.load(source)


def is_active(record):
    return record.get("status") == "active"


def format_customer(record):
    return f"{record.get('id')} - {record.get('name')} ({record.get('email')})"


try:
    records = load_customers(INPUT_FILE)
    active = [format_customer(r) for r in records if is_active(r)]
    report = ["Active Customer Report", "======================", f"Total active customers: {len(active)}", ""] + active
    os.makedirs(os.path.dirname(OUTPUT_FILE), exist_ok=True)
    with open(OUTPUT_FILE, "w", encoding="utf-8") as out_file:
        out_file.write("\n".join(report))
    print("Report created successfully.")
except FileNotFoundError:
    print("The input file was not found. Please check the path.")
except json.JSONDecodeError:
    print("The input file contains invalid JSON.")
```
