```python
# data of male who play cricket
male_play_cricket = df[(df['gender'] == 'male') & (df['play_circket'] == 'yes')]
male_play_cricket
```

```python
# data of male who  not play cricket
male_not_play_cricket = df[(df['gender'] == 'male') & (df['play_circket'] == 'no')]
male_not_play_cricket
```

```python
#probalites of yes/no playing cricket
p_yes = len(df[df['play_circket'] == 'yes']) / len(df)
p_yes
p_no = len(df[df['play_circket'] == 'no']) / len(df)
p_no
```

```python
# entrophy of male playing cricket
entropy_male_play_cricket = - (p_male_play_cricket * np.log2(p_male_play_cricket) + p_male_not_play_cricket * np.log2(p_male_not_play_cricket))
entropy_male_play_cricket   
```

```python
male = df[df["gender"] == "male"]["play_circket"]
female = df[df["gender"] == "female"]["play_circket"]
```

```python
male_counts = male.value_counts()

p_yes = male_counts["yes"] / len(male)
p_no = male_counts["no"] / len(male)

male_entropy = - (p_yes * np.log2(p_yes) + p_no * np.log2(p_no))

male_entropy
```

```python
female_counts = female.value_counts()

p_yes = female_counts["yes"] / len(female)
p_no = female_counts["no"] / len(female)

female_entropy = - (p_yes * np.log2(p_yes) + p_no * np.log2(p_no))

female_entropy
```

```python
overall_counts = df["play_circket"].value_counts()

p_yes = overall_counts["yes"] / len(df)
p_no = overall_counts["no"] / len(df)

overall_entropy = - (p_yes * np.log2(p_yes) + p_no * np.log2(p_no))

overall_entropy
```

```python
weighted_entropy = (
    (len(male) / len(df)) * male_entropy +
    (len(female) / len(df)) * female_entropy
)

weighted_entropy
```
