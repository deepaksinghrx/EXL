```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
```

```python
# dummy data for demonstration PCA
data=np.array([[2.5, 2.4], [0, 0], [0, 0]])
```

```python
np.random.seed(42)
n_samples = 50

z1 = np.random.normal(0, 1, n_samples)
z2 = np.random.normal(0, 1, n_samples)

X = pd.DataFrame(
    {
        "feature_1": 2 * z1 + 0.5 * z2 + np.random.normal(0, 0.3, n_samples),
        "feature_2": 1.5 * z1 - 0.7 * z2 + np.random.normal(0, 0.4, n_samples),
        "feature_3": -0.8 * z1 + 1.8 * z2 + np.random.normal(0, 0.5, n_samples),
        "feature_4": 1.2 * z1 + 1.2 * z2 + np.random.normal(0, 0.4, n_samples),
        "feature_5": 0.4 * z1 - 0.2 * z2 + np.random.normal(0, 0.6, n_samples),
    }
)

X["group"] = np.where(z1 + z2 > 0, "A", "B")

print(X.head())
```
