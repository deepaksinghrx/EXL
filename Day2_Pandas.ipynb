```python
# csv report for age group wise summary statistics
# age_grp | p_id_count | total_cost_mean| total_cost_sum | visit_mean
# young(0-30) |3  | 111.56 | 56454.23 | 5.2


group_age=df_patients['age'].apply(lambda x: 'young' 
                                    if x<=30 
                                    else ('middle age' 
                                                   if x<=50 
                                                        else 'senior'))
df_patients['age_group']=group_age
df_patients
```

```python
report=df_patients.groupby('age_group').agg(
    patient_count=('patient_id','count'),
    cost_mean=('total_cost','mean'),
    cost_sum=('total_cost','sum'),
    visit_mean=('visits','mean')
)
```

```python
report
```

```python
# condn summaary
# p_id_count, age_mean, max_age, TOtalCOst_mean, visit _mean
second_report=df_patients.groupby('condition').agg(
    patient_count=('patient_id','count'),
    age_mean=('age','mean'),
    age_max=('age','max'),
    cost_mean=('total_cost','mean'),
    visit_mean=('visits','mean')
)
second_report
```

```python
report.to_csv('age_group_report.csv')
second_report.to_csv('condition_report.csv')

```

```python
import matplotlib.pyplot as plt

plt.bar(report.index, report['cost_mean'])
plt.xlabel('Age Group')
plt.ylabel('Average Cost')
plt.title('Average Cost by Age Group')
plt.show()
```
