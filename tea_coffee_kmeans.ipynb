```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python


data = {
    "Age": [18, 22, 25, 28, 30, 32, 35, 40, 45, 20, 23, 27, 38, 42, 50],
    "Work_Hours": [4, 5, 6, 7, 8, 9, 10, 9, 8, 5, 6, 7, 10, 9, 8],
    "Sleep_Hours": [8, 7, 8, 7, 6, 6, 5, 5, 6, 8, 7, 7, 5, 5, 6],
    "Cups_Per_Day": [2, 3, 2, 3, 4, 5, 6, 5, 4, 2, 3, 3, 6, 5, 4],
    "Preference": [
        "Tea", "Tea", "Tea", "Tea",
        "Coffee", "Coffee", "Coffee", "Coffee", "Coffee",
        "Tea", "Tea", "Tea",
        "Coffee", "Coffee", "Coffee"
    ]
}

df = pd.DataFrame(data)


```
