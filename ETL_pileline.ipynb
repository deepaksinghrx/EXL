```python
class pipeline:
    def __init__(self, file_path: str, output_dir: str = 'output') -> None:
        self.file_path = Path(file_path) if Path(file_path).exists() else Path('Datasets') / file_path
        self.output_dir = Path(output_dir) if Path(output_dir).exists() else Path('Datasets') / output_dir
        self.output_dir.mkdir(exist_ok=True)

    def load(self) -> pd.DataFrame:
        try:
            df = pd.read_csv(self.file_path)
            print("Data loaded")
            return df
        except Exception as e:
            print("Error loading data")
            return pd.DataFrame()

    def transform(self, df: pd.DataFrame) -> pd.DataFrame:
        df['cost_per_visit'] = df['cost'] / df['visit']
        age_bins = [0, 18, 35, 50, 65, np.inf]
        age_labels = ["young", "adult", "middle-aged", "senior", "elderly"]
        df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)
        df['risk'] = pd.cut(df['visit'], bins=[0, 1, 3, np.inf], labels=['low', 'medium', 'high'])
        print("Data transformed")
        return df

    def export(self, df: pd.DataFrame, filename: str) -> None:
        output_path = self.output_dir / filename
        try:
            df.to_csv(output_path, index=False)
            print(f"Data exported to {output_path}")
        except Exception as e:
            print(f"Error exporting data: {e}")


pipeline = pipeline('Datasets/health_dataset.csv')
raw_df = pipeline.load()    
prepared_df = pipeline.transform(raw_df)    
pipeline.export(prepared_df, 'transformed_health_data.csv')

prepared_df.head()
```
