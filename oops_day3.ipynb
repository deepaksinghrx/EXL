```python
import pandas as pd
import numpy as np

file_path = 'Datasets/ecommerce_dataset.csv'

def load_data(file_path):
    try:
        data = pd.read_csv(file_path)
        return data
    except Exception as e:
        print(f"An error occurred: {e}")
        return None
    
def preprocess_data(data):
    data=data.drop_duplicates()
    data['customer_age']=data['customer_age'].fillna(data['customer_age'].median())
    return data

def filter_data(data,payment_method):
    return data[data['payment_method'] == payment_method]

def export_data(data,file_path):
    data.to_csv(file_path,index=False)

if __name__ == "__main__":
    data = load_data(file_path)
    preprocess = preprocess_data(data)
    filtered_data = filter_data(preprocess, 'Credit Card')
    export_data(filtered_data, 'filtered_ecommerce_dataset.csv')

```

```python
class datapipeline:
    def __init__(self, file_path: str):
        self.file_path = file_path
        self.data = None
        self.cleaned_data = False

    def load_data(self) -> 'datapipeline':
        try:
            self.data = pd.read_csv(self.file_path)
            return self
        except Exception as e:
            print(f"An error occurred: {e}")
            self.data = None

    def preprocess_data(self) -> 'datapipeline':
        self.data=self.data.drop_duplicates()
        self.data['customer_age']=self.data['customer_age'].fillna(self.data['customer_age'].median())
        self.cleaned_data = True
        return self

    def export_data(self, out: str) -> None:
        if not self.cleaned_data:
            raise RuntimeError("Data has not been cleaned first")
        self.data.to_csv(out, index=False)
```

```python
# inhenritance example
class Patient:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def get_info(self):
        return f"Patient Name: {self.name}, Age: {self.age}"
    
class InPatient(Patient):
    def __init__(self, name, age, room_number):
        super().__init__(name, age)
        self.room_number = room_number

    def get_info(self):
        return f"Patient Name: {self.name}, Age: {self.age}, Room Number: {self.room_number}"
    
class OutPatient(Patient):
    def __init__(self, name, age, appointment_date):
        super().__init__(name, age)
        self.appointment_date = appointment_date

    def get_info(self):
        return f"Patient Name: {self.name}, Age: {self.age}, Appointment Date: {self.appointment_date}"
    
in_patient = InPatient("John Doe", 45, "101A")
out_patient = OutPatient("Jane Smith", 30, "2023-07-15")
print(in_patient.get_info())
print(out_patient.get_info())



```

```python
# ------------------------------------------------------------------
# Cell 7: Typed DataPipeline lab
# ------------------------------------------------------------------
# This class is the real pipeline example for Day 3.
# It combines: load -> transform -> export.

import sys
from pathlib import Path
class TypedDataPipeline:
    # __init__ prepares the file paths for the pipeline.
    def __init__(self, file_path: str, output_dir: str = 'output') -> None:
        # Use the given path if it exists.
        self.file_path = Path(file_path) if Path(file_path).exists() else Path('Datasets') / file_path
        # Use the given output folder if it exists; otherwise create the Day3 version.
        self.output_dir = Path(output_dir) if Path(output_dir).exists() else Path('Datasets') / output_dir
        # Make sure the output folder exists before writing files.
        self.output_dir.mkdir(exist_ok=True)

    # load() reads the raw CSV dataset into memory.
    def load(self) -> pd.DataFrame:
        return pd.read_csv(self.file_path)

    # transform() adds new business-friendly columns to the data.
    def transform(self, df: pd.DataFrame) -> pd.DataFrame:
        transformed = df.copy()
        # Compute average cost per visit for each patient.
        transformed['cost_per_visit'] = transformed['cost'] / transformed['visit']
        # Group ages into meaningful categories.
        transformed['age_group'] = pd.cut(
            transformed['age'],
            bins=[0, 30, 45, 60, 100],
            labels=['Young', 'Adult', 'Senior', 'Elderly'],
            right=True,
        )
        return transformed

    # export() saves the transformed data to a CSV file.
    def export(self, df: pd.DataFrame, filename: str) -> None:
        output_file = self.output_dir / filename
        df.to_csv(output_file, index=False)
        print(f'Exported to {output_file}')


# Create the pipeline object.
pipeline = TypedDataPipeline('data/health_dataset.csv')

# Step 1: load the raw data.
raw_df = pipeline.load()

# Step 2: transform the data with new columns.
prepared_df = pipeline.transform(raw_df)

# Step 3: export the transformed data to a file.
pipeline.export(prepared_df, 'healthcare_pipeline_output.csv')

# Show the first rows of the transformed table.
prepared_df.head()
```
