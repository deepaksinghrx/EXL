```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

X = df.drop('c_id', axis=1)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Elbow method
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('K')
plt.ylabel('WCSS')
plt.show()


kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
labels = kmeans.fit_predict(X_scaled)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
plt.scatter(X_pca[:,0], X_pca[:,1], c=labels)
plt.title('Clusters')
plt.show()
```

```python
features=['tea', 'coffee', 'greentea', 'cold_drink', 'gym_days',
       'water_ltres', 'work_hrs', 'netflix_hrs']
```

```python
x=df[features]
```

```python
x
```

```python
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(x)
```

```python
k=KMeans(n_clusters=4,random_state=42,n_init=10)
```

```python
df["cluster"]=k.fit_predict(X_scaled)
```

```python
df
```

```python
df.groupby("cluster")[features].mean()
```

```python
df.columns
```

```python
drinking_columns=['tea', 'coffee', 'greentea', 'cold_drink']
```

```python
cluster_profiles=df.groupby("cluster")[drinking_columns].mean()
```

```python
cluster_profiles
```

```python
dominant_drink=cluster_profiles.idxmax(axis=1)
```

```python
dominant_drink
```

```python
type(dominant_drink)
```

```python
dominant_drink.items
```

```python
label_mapping={}
```

```python
for cluster_id,drink in dominant_drink.items():
    if drink=="tea":
        label_mapping[cluster_id]="Tea Lover"
    elif drink=="coffee":
        label_mapping[cluster_id]="Coffee Lover"
    elif drink=="greentea":
        label_mapping[cluster_id]="Health Lover"
    elif drink=="cold_drink":
        label_mapping[cluster_id]="cold_drink Lover"
```

```python
df["segment"]=df["cluster"].map(label_mapping)
```
